Para este código se utiliza la biblioteca Matplotlib de Python para crear un grafico. Primero, se crea una figura y un conjunto de ejes utilizando la función subplots(). Luego, se definen cuatro listas vacías: x_onion, y_onion, x_soda, y_soda.
Se hacen dos bucles for para recorrer do diccionarios diferentes: lis_onion y lis_soda, estas contienen una clave "loc" que es la localización, la cual contiene una tupla con las coordenadas (x,y) que representan la ubicación de item. En cada iteración del primer bucle, se agrega el valor de la coordenada x de la ubicación del item actual en lis_onion a la lista x_onion y se agrega el valor de la coordenada y a la lista y_onion. De esta manera similar, en cada iteración del segundo bucle, se agregan los valores de las coordenadas x e y de la ubicación del item en list_soda a las listas correspondientes. En resumen, se utilizara para extraer las coordenadas de los elementos en dos listas separadas para su posterior uso en un grafico. Luego, se utiliza la función scatter() de Matplotlib para agregar puntos al grafico. Se agregan dos conjuntos de puntos, asociando cada conjunto con las listas creadas anteriormente, la lista x_onion y y_onion con los puntos color verde y la lista x_soda y y_soda para los puntos de color rojo. Además, se emplea la biblioteca oxmnx para crear un mapa que represente el municipio de Guanabacoa, con una distancia de 2000 m; se agrega la función plot_graph() de oxmnxpara agregar el mapa al gráfico al existente, con nodos sin color.

```python
#Se utiliza la biblioteca de Matplotlib para crear un grafico.
fig, ax = plt.subplots() 
x_onion = []
y_onion = []
x_coffe = []
y_coffe = []

#Recorrer la lista y agregar el valor de la coordenada x a la lista x_onion y de la coordenada y a la lista y_onion
for item in lis_onion:
    x_onion.append(item["loc"][0])
    y_onion.append(item["loc"][1])

#Recorrer la lista y agregar el valor de la coordenada x a la lista x_coffe y de la coordenada y a la y_coffe  
for item in list_soda:
    x_coffe.append(item["loc"][0])
    y_coffe.append(item["loc"][1])

#Generar el grafico
ax.scatter(y_coffe, x_coffe, c='red', s=20)  
ax.scatter(y_onion, x_onion, c='green', s=20)
map = ox.graph_from_address("Guanabacoa,Cuba", dist=2000)
ox.plot_graph(map, ax=ax, node_color="None")
plt.show()
```

Este mapa se creó con el fin de visualizar todo el territorio que fue recorrido de Guanabacoa, utilizando dos colores para representar los tipos de lugares, en este caso se escogió el rojo para las cafeterías y el verde para los agros.
Se puede decir que los lugares escogidos están casi todos concentrados en una sola

A continuación, se aplicara el método ".groupby()" para agrupar los datos, en este caso analizaremos el envase y su color ,y nos auxiliaremos del método ".count()" para contar la cantidad de marcas que utilizan envases similares. Antes de ponerte a analizar la gráfica de barra, primeramente investiguemos un poco acerca de la diferencia que existe entre la cerveza en lata o en botella.
Básicamente, lo que ocurre es que la botella, al tener un cuello muy estrecho, permite que se pueda desplazar fácilmente el oxigeno y que se pueda taponear rápido sin que haya presencia de este "enemigo" de la cerveza. Por el contrario, la lata tiene una abertura muy grande y esto dificulta mucho el momento de taponearla en ausencia absoluta de oxígeno. Esto provoca que el proceso de oxidación de la cerveza en lata sea mucho mas rápido, podemos decir hasta el momento que la botella conserva un mejor sabor al paladar que la lata. Ahora bien, lo que lleva a las cerveceras a seguir utilizando un formato que conserva el producto peor que el vidrio, es la comodidad a la hora de transportarla, apilarla y almacenarla. En cuanto a la temperatura y la luz , las botellas dejan pasar demasiada luz y las latas no dejan pasar nada. Además la lata se calienta y se enfría más rápido que la botella esto dependiendo de lo queramos en cada momento, puede ser bueno o malo, y ahora tú, que prefieres,  lata o botella?
Esta información fue sacada de una publicación de: https://www.directoalpaladar.com/cata-de-cerveza/cerveza-mejor-lata-botella-vidrio
Ahora volviendo al análisis de la grafica de barra se puede observar que predomina la lata, y además el color verde. Y es que si miramos bien la grafica de abajo de: Mayores exportadores de cerveza por continente, casi toda la cerveza importada es de Europa, esto responde, a porque casi toda la cerveza que se vende en Guanabacoa es en lata. Ya que con la información anterior, lo atractivo de la lata es la comodidad para transportarla, apilarla y almacenarla.

Se crea un diccionario que contendrá a los continentes con su frecuencia asociada. Luego se cambia la frecuencia por el porcentaje que representa del total de datos que se recogieron y por último se representará en un gráfico de pastel.

```python
#Recorrer la lista lis_beer y agregar los elementos al diccionario
dicti_continents = {} 
for item in lis_beer: 
    if item["continente"] in dicti_continents: 
        dicti_continents[item["continente"]] += 1
    else:
        dicti_continents[item["continente"]] = 1

print(dicti_continents)

#Generar el grafico
plt.pie(dicti_continents.values(), labels = dicti_continents.keys(), autopct="%0.1f%%")
plt.title("Mayores exportadores de cerveza por continente")
plt.show()
```

#### ¿De que país es la cerveza que más está distríbuida por las cafeterías de Guanabacoa?

```python
#Recorrer la lista lis_beer y agregar los elementos al diccionario
dicti_countries = {}
for item in lis_beer:
    if item["pais_de_procedencia"] in dicti_countries:
        dicti_countries[item["pais_de_procedencia"]] += 1
    else:
        dicti_countries[item["pais_de_procedencia"]] = 1

print(dicti_countries)

#Recorrer el diccicionario y agregar los valores al nuevo diccionario
aux ={}
other = 0
for country in dicti_countries:
    if dicti_countries[country] > 5.0:
        aux[country] = dicti_countries[country]
    else:
        other += dicti_countries[country]
aux["Other"] = other
dicti_countries = aux

#Generar el grafico
plt.pie(dicti_countries.values(), labels = dicti_countries.keys(), autopct="%0.1f%%")
center_circle= plt.Circle((0,0), 0.40, fc='white')
# Selecciona la figura, selecciona los ejes y annade el elemento circulo en el grafico
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.title("Mayores exportadores de cerveza por paises")
plt.show()
```
